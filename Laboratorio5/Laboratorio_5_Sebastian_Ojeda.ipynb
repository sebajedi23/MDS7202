## Análisis por Nivel Educacional Etnia de los Padres [1.5 Puntos]

El rector, basado en su experiencia, cree fuertemente que el nivel educacional y la etnia de los padres influyen en las notas que obtienen sus hijos.
Como científicos de datos, ud. y su equipo creen que deben encontrar evidencia para confirmar o refutar la hipótesis del rector.

### Tabla de Resumen

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión:

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

### 1.3.4 Información por Categorías [1 punto]

````python
#Calculamos los promedios y los almacenamos en una variable GPA
df_grades['GPA'] = df_grades[['math score', 'reading score', 'writing score']].mean(axis=1)

#Hacemos el mapeo de la varibale parental level of education
parental_map = {
    'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    "bachelor's degree": 'college',
    "associate's degree": 'college',
    "master's degree": 'postgraduate'
}

#Aplicamos el mapeo
df_grades['simple parental level of education'] = df_grades['parental level of education'].map(parental_map)

#Agrupamos por raza/etnia y la variable recien mapeada
result = df_grades.groupby(['race/ethnicity', 'simple parental level of education'])[
    ['math score', 'reading score', 'writing score', 'GPA']].agg({
    'math score': 'mean',
    'reading score': 'mean',
    'writing score': 'mean',
    'GPA': 'mean'
}).reset_index()

#Agregamos el total por grupo
count_df = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')
result['count'] = count_df['count']

#Agregamos el porcentaje
total_students = len(df_grades)
result['percentage'] = (result['count'] / total_students * 100).apply(lambda x: f'{x:.2f}%')

#Mostramos los resultados con 2 decimales
result = result.round(2)
result
````

````python
#Unimos usando inner join
merged_inner = df_grades.merge(other_grades, how='inner')
merged_inner
````

```
De acuerdo a lo visto, se puede decir que el outer join toma todos los valores de ambos dataframes y en caso de no haber un match en
alguna fila, la añade al final rellenando con NaN los valores que no se encontraron debido a que la nueva base de datos no los contiene.
Esto se puede ver reflejado en las 1000 filas que tiene, correspondientes a las 1000 de la nueva base de datos, por lo que se
confirma que se agregaron nuevas filas a las 875 originales.

El caso de left join toma todos los valores del primer dataframe y los une con los del segundo dataframe, retornando únicamente las
filas del primer dataframe, las filas del segundo dataframe que no tienen coincidencias se desestiman. Debido a esto es que el
dataframe resultante tiene las mismas 875 filas del dataframe original.

Para el right join es similar al anterior pero al revés, es decir, en lugar de retornarse las 875 filas del primer dataframe, se
retornan las 1000 filas del segundo aún cuando no tengan coincidencias con el otro datafraem. Por esto mismo, este dataframe
tiene 1000 filas. Usando .describe() es posible notar que todas las filas exceptuando las nuevas, tienen sólo 875 valores, pues los no
encontrados se reemplazan por NaN.

Por último el inner join lo que hace es devolver únicamente las filas que tienen coincidencia en ambos dataframes. Este dataframe
tiene 875 filas, lo que quiere decir que el segundo dataframe contiene exactamente los mismos 875 alumnos del primer dataframe,
más algunos adicionales. En caso de que el segundo dataframe tuviera 1000 alumnos que no están en el primer dataframe, se obtendría un
dataframe vacío, pues no habría coincidencias.

Dado lo anteriormente mencionado, se utilizará el inner join, ya que de esta forma siempre nos aseguramos de que no nos falten datos
en ninguna de las columnas.
```
