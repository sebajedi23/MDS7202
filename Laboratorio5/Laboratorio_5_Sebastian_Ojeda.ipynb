## An치lisis por Nivel Educacional Etnia de los Padres [1.5 Puntos]

El rector, basado en su experiencia, cree fuertemente que el nivel educacional y la etnia de los padres influyen en las notas que obtienen sus hijos.
Como cient칤ficos de datos, ud. y su equipo creen que deben encontrar evidencia para confirmar o refutar la hip칩tesis del rector.

### Tabla de Resumen

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificaci칩n a trav칠s de un mapeo (investigar el m칠todo `map()`) de la variable `parental level of education`  seg칰n la siguiente conversi칩n:

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.


Utilizar la tabla de resultados esperados como gu칤a para desarrollar este punto.

### 1.3.4 Informaci칩n por Categor칤as [1 punto]

````python
#Calculamos los promedios y los almacenamos en una variable GPA
df_grades['GPA'] = df_grades[['math score', 'reading score', 'writing score']].mean(axis=1)

#Hacemos el mapeo de la varibale parental level of education
parental_map = {
    'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    "bachelor's degree": 'college',
    "associate's degree": 'college',
    "master's degree": 'postgraduate'
}

#Aplicamos el mapeo
df_grades['simple parental level of education'] = df_grades['parental level of education'].map(parental_map)

#Agrupamos por raza/etnia y la variable recien mapeada
result = df_grades.groupby(['race/ethnicity', 'simple parental level of education'])[
    ['math score', 'reading score', 'writing score', 'GPA']].agg({
    'math score': 'mean',
    'reading score': 'mean',
    'writing score': 'mean',
    'GPA': 'mean'
}).reset_index()

#Agregamos el total por grupo
count_df = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')
result['count'] = count_df['count']

#Agregamos el porcentaje
total_students = len(df_grades)
result['percentage'] = (result['count'] / total_students * 100).apply(lambda x: f'{x:.2f}%')

#Mostramos los resultados con 2 decimales
result = result.round(2)
result
````
