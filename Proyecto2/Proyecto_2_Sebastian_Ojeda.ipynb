## 3.1 Preprocesamiento con ColumnTransformer

Para comenzar, a partir de lo discutido en la sección anterior notamos que variables como "occupation", "payment_of_min_amount" y "payment_behaviour" tienen pocos valores únicos y al revisar los mismos es posible observar que estos corresponden a diversas categorías a las que puede pertenecer un cliente en distintos ámbitos, por lo mismo, se consideran como categóricas.

Luego, las columnas "customer_id", "monthly_inhand_salary", "credit_history_age", "num_of_delayed_payment" no serán utilizadas, la primera por ser un identificador que no aporta más información y las restantes debido a la gran cantidad de valores nulos que poseen.

Finalmente, "credit_score" corresponde a la variable objetivo, y todas las variables restantes se clasifican como numéricas luego de decidirse que ninguna de estas calza como ordinal.

```python
#Definimos listas para cada tipo de variable
categorical = ["occupation", "payment_of_min_amount", "payment_behaviour"]

target = "credit_score"

numeric = ["num_bank_accounts", "num_of_loan", "num_credit_inquiries",
           "total_emi_per_month", "credit_utilization_ratio",
           "changed_credit_limit", "monthly_balance", "num_credit_card",
           "interest_rate", "age", "annual_income", "delay_from_due_date",
           "outstanding_debt", "amount_invested_monthly"]

unused = ["customer_id", "monthly_inhand_salary", "credit_history_age",
          "num_of_delayed_payment"]
```

Posteriormente, nos aseguramos de que las columnas tratadas como variables numéricas tengan asignado correctamente su tipo (int 8, int16, int64 o float), manejando posibles errores de lectura y optimizando el uso de memoria al convertir a tipos de dato entero cuando sea posible.\
De la misma forma trabajamos las columnas categóricas y fijamos su tipo como category para corregir posibles errores de lectura.

```python
#Convertimos los tipos de dato en las columnas numéricas
for column in numeric:
    df[column] = pd.to_numeric(df[column], errors='coerce', downcast='integer')

#Convertimos los tipos de dato en las columnas categóricas
for col in categorical:
    df[col] = df[col].astype('category')
```

```python
#Revisamos los tipos de dato
df.dtypes
```

Con esto se puede ver que columnas que se trataban como float pero poseían únicamente valores enteros ahora son tratados como int16 o int64 según sea más conveniente.

```python
#Creamos el objeto ColumnTransformer con las transformaciones para cada tipo de variable
ct = ColumnTransformer([('OneHot', OneHotEncoder(sparse=False, handle_unknown="ignore",drop='if_binary'), categorical),
                        ('RobustScaler', RobustScaler(), numeric)
                        ],
                        remainder='passthrough')

#Definimos que la salida mantenga el formato pandas
ct.set_output(transform="pandas")
```

*   "changed_credit_limit": Un valor negativo puede indicar que el límite de cupo de la tarjeta de crédito fue reducido, por lo que también tiene sentido.
*   "num_of_delayed_payment": Un valor negativo en esta variable no tiene sentido. De todos modos esta fue seleccionada como una de las que no se utilizará, por lo que no es necesario trabajarlas.
*   "monthly_balance": Un balance negativo es posible, por lo que también tiene sentido este tipo de valores en esta variable.

De este modo las columnas "age" y "num_of_delayed_payment" son las únicas a las que se debe trabajar sus valores negativos.\
Adicionalmente la columna "age", además de valores negativos, posee valores positivos sin sentido, incluso llegando a los miles de años, por lo que se limitarán sus valores hasta 110, ya que se considera que esta es una edad de rara ocurrencia pero de todos posible. Por otro lado, el mínimo valor positivo es de 14, por lo que se tomará el valor 10 como límite inferior al transformar los datos negativos a fin de no eliminar los extremos inferiores y dejar un poco de holgura a posibles valores inferiores a estos.

A continuación se definirá una función que aplique las transformaciones previamente explicadas.

```python
#Definimos una función para aplicar las transformaciones discutidas anteriormente
def preprocess(df):
    """Procesa y transforma un DataFrame de datos.

    Parameters
    ----------
    df : pandas.DataFrame
        El DataFrame de entrada que se va a procesar.

    Returns
    -------
    pandas.DataFrame
        Un nuevo DataFrame después de aplicar las debidas transformaciones.

    Summary
    -------
    Esta función toma un DataFrame `df` y realiza las siguientes operaciones:
    1. Crea una copia del DataFrame original.
    2. Transforma los valores negativos a positivos en las columnas "age" y
       "num_of_delayed_payment".
    3. Establece los valores de "age" como enteros y ajusta los valores de
       "num_of_delayed_payment" a positivos.
    4. Define límites inferior y superior en la columna "age" (10 y 110,
       respectivamente).
    5. Elimina columnas especificadas como no utilizadas ("customer_id",
       "monthly_inhand_salary", "credit_history_age", "num_of_delayed_payment").
    6. Imputa valores faltantes en las columnas "amount_invested_monthly",
       "monthly_balance", "num_credit_inquiries", "changed_credit_limit"
       utilizando la media de cada columna.
    """
    #Creamos una copia de los datos
    df2 = df.copy()

    #Transformamos los valores negativos a positivos en "age" y "num_of_delayed_payment"
    df2['age'] = df2['age'].abs()
    df2['age'] = df2['age'].astype('int')
    df2['num_of_delayed_payment'] = df2['num_of_delayed_payment'].abs()

    #Definimos los límites inferior y superior en "age" como 10 y 110
    df2['age'] = df2['age'].apply(lambda x: max(10, min(110, x)))

    #Definimos las variables que no se usarán
    unused = ["customer_id", "monthly_inhand_salary", "credit_history_age",
              "num_of_delayed_payment"]

    #Eliminamos las variables que no se usarán
    for col in unused:
        if col in df2.columns:
            df2 = df2.drop(columns=col)

    #Definimos las columnas a imputar
    to_impute = ["amount_invested_monthly", "monthly_balance", "num_credit_inquiries", "changed_credit_limit"]

    #Aplicamos la media de la columna como método de imputación
    imputer = SimpleImputer(strategy='mean')
    df2[to_impute] = imputer.fit_transform(df2[to_impute])

    #Retornamos el DataFrame ya trabajado
    return df2
```

```python
#Definimos un FunctionTransformer para la función anterior
preprocessing = FunctionTransformer(preprocess)
```

```python
def create_pipeline(DataProcessor,ColTransformer, model):
    return Pipeline(steps=[
        ("preprocessing", DataProcessor),
        ("ct", ColTransformer),
        ("model", model)
    ])

seed = 42  # Seed para reproducibilidad

clf_dummy = create_pipeline(preprocessing, ct, DummyClassifier(strategy='stratified', random_state=seed))
clf_logistic = create_pipeline(preprocessing, ct, LogisticRegression(random_state=seed))
clf_knn = create_pipeline(preprocessing, ct, KNeighborsClassifier())
clf_tree = create_pipeline(preprocessing, ct, DecisionTreeClassifier(random_state=seed))
clf_svc = create_pipeline(preprocessing, ct, SVC(random_state=seed))
clf_rf = create_pipeline(preprocessing, ct, RandomForestClassifier(random_state=seed))
clf_lgbm = create_pipeline(preprocessing, ct, LGBMClassifier(random_state=seed))
clf_xgb = create_pipeline(preprocessing, ct, XGBClassifier(random_state=seed))
```

```python
#Creamos un DataFrame para almacenar las métricas obtenidas con cada modelo optimizado
metrics_opt = pd.DataFrame(columns=["model", "precision val", "precision val (opt)"])
```

```python
def objective_lgbm(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 5000)
    max_depth = trial.suggest_int("max_depth", 2, 10)
    learning_rate = trial.suggest_loguniform("learning_rate", 1e-3, 1e-1)

    model = LGBMClassifier(verbose=-1, random_state=42, n_estimators=n_estimators, max_depth=max_depth, learning_rate=learning_rate)
    pipeline = ImbPipeline(steps=[("preprocessing", preprocessing),
                                  ("ct", ct),
                                  ("balance", SMOTE()),
                                  ("model", model)
                                  ])

    # Entrenamiento y evaluación
    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_val)
    score = precision_score(y_val, preds)

    return score

# Crear y ejecutar el estudio de optimización para LightGBM
study_lgbm = optuna.create_study(direction="maximize")
study_lgbm.optimize(objective_lgbm, timeout = 5*60)

print("Mejores parámetros para LightGBM:", study_lgbm.best_params)
```

```python
def objective_xgb(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 5000)
    max_depth = trial.suggest_int("max_depth", 2, 10)
    learning_rate = trial.suggest_loguniform("learning_rate", 0.005, 0.5)
    min_child_weight = trial.suggest_float("min_child_weight", 1, 10)
    subsample = trial.suggest_float("subsample", 0.5, 1.0)
    colsample_bytree = trial.suggest_float("colsample_bytree", 0.5, 1.0)
    reg_alpha = trial.suggest_float("reg_alpha", 0.0, 1.0)
    reg_lambda = trial.suggest_float("reg_lambda", 0.0, 1.0)

    model = XGBClassifier(
        random_state=42,
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        min_child_weight=min_child_weight,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        reg_alpha=reg_alpha,
        reg_lambda=reg_lambda,
        n_jobs=-1
    )

    pipeline = ImbPipeline(steps=[("preprocessing", preprocessing),
                                  ("ct", ct),
                                  ("balance", SMOTE()),  # Balanceo de clases
                                  ("model", model)
                                  ])

    # Entrenamiento y evaluación
    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_val)
    score = precision_score(y_val, preds)

    return score

# Crear y ejecutar el estudio de optimización para XGBoost
study_xgb = optuna.create_study(direction="maximize")
study_xgb.optimize(objective_xgb, timeout=5*60)

print("Mejores parámetros para XGBoost:", study_xgb.best_params)
```

```python
#Obtenemos las métricas para ambos modelos optimizados
best_score_lgbm = study_lgbm.best_value
best_score_xgb = study_xgb.best_value

#Agregamos las métricas obtenidas al DataFrame
metrics_opt = metrics_opt.append({"model": "LightGBM", "precision val": precision_val_lgbm, "precision val (opt)": best_score_lgbm}, ignore_index=True)
metrics_opt = metrics_opt.append({"model": "XGBoost", "precision val": precision_val_xgb, "precision val (opt)": best_score_xgb}, ignore_index=True)
metrics_opt
```
