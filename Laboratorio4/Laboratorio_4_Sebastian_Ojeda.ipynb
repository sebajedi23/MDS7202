````python
def exploratory_data_analysis(dataframe):
    """Recibe un dataframe y entrega 9 puntos relevantes para su análisis

    Parameters
    ----------
    dataframe : pandas.core.frame.DataFrame
        Dataframe de pandas a analizar
    """
    # 1. Imprimir un mensaje con las dimensiones del Dataframe.
    print(f"1.- El DataFrame tiene {dataframe.shape[0]} filas y {dataframe.shape[1]} columnas\n")

    # 2. Imprimir un mensaje con los nombres de las columnas que conforman el
    # DataFrame.
    print(f"2.- El DataFrame está compuesto por las siguientes columnas: {(dataframe.columns.values)}\n")

    # 3. Mostrar los primeros 5, los últimos 5 elementos del DataFrame y realice
    # un muestreo de 5 elementos de forma aleatoria.
    print("3.- Ejemplos de filas del DataFrame:")
    print("\nPrimeras 5 filas:")
    display(df_grades.head())
    print("\nÚltimas 5 filas:")
    display(df_grades.tail())
    print("\nMuestreo aleatorio de 5 filas:")
    display(dataframe.sample(5))
    print("\n")

    # 4. Descripción numérica del DataFrame
    print("4.- Descripción numérica del Dataframe:")
    display(dataframe.describe())
    print("\n")

    # 5. Cantidad de valores nulos por columna
    print("5.- Cantidad de valores nulos por columna:")
    display(dataframe.isna().sum())
    print("\n")

    # 6. Cantidad de valores únicos por columna
    print("6.- Cantidad de valores únicos por columna:")
    display(dataframe.nunique())
    print("\n")

    # 7. Presenta un total de filas con nombres duplicados
    a = int(df_grades.duplicated(subset="names").sum())
    print(f"7.- Presenta un total de {a} filas con nombres duplicados.\n")

    # 8. Las filas con nombres duplicados
    print("8.- Las filas con nombres duplicados son:")
    display(dataframe[dataframe.duplicated(subset="names")])
    print("\n")

    # 9. Boxplot de notas utilizando Plotly Express
    print('9.- Boxplot de Notas')
    # Crear un DataFrame con las tres columnas a representar
    data = dataframe[["math score", "reading score", "writing score"]]

    # Derretir el DataFrame para tener una sola columna "score" y una columna "subject" para diferenciar las materias
    melted_data = pd.melt(data, var_name="subject", value_name="score")

    # Mostrar el boxplot de las notas, usando plotly
    fig = px.box(melted_data, y="score", color="subject")

    # Mostramos el boxplot
    fig.show()
````

- [ ] Comprobar que no existen anomalías en los datos a traves de la función de la sección 1.1.

**Hint**

Puede ser útil los métodos ```.drop_duplicates()``` y ```.dropna()``` como también utilizar queries con máscaras booleanas.


**Respuesta:**

````python
def clean_data(df):
    """Limpia un dataframe de entradas repetidas o valores inexistentes o fuera
    de la escala de notas

    Parameters
    ----------
    df : pandas.core.frame.DataFrame
        Dataframe a limpiar
    """

    #Primero eliminamos las notas fuera de la escala
    df = df[(df["math score"] >= 0) & (df["math score"] <= 100)]
    df = df[(df["reading score"] >= 0) & (df["reading score"] <= 100)]
    df = df[(df["writing score"] >= 0) & (df["writing score"] <= 100)]

    #Luego eliminamos las entradas duplicadas por nombre
    df = df.drop_duplicates(subset="names")

    #Eliminamos las entradas nulas
    df = df.dropna()

    #Finalmente reseteamos los índices para que no haya saltos debido a las
    #filas eliminadas
    df = df.reset_index(drop=True)

    return df
````

### 1.3.1 Transformación de notas a "Nota Chilena" [0.5 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
    """Convierte las notas de una columna de escala estadounidense a chilena
    redondeadas al segundo decimal

    Parameters
    ----------
    dataframe_in : pandas.core.frame.DataFrame
        Dataframe de pandas al que se quiere transformar las notas
    columna : str
        Nombre de la columna del dataframe a transformar
    """
    #Usamos la fórmula dada para convertir las notas de la columna en cuestión
    dataframe_in[columna] = dataframe_in[columna].apply(lambda x: round(x * 0.06 + 1, 2))
    return dataframe_in
````

````python
#Aplicamos la función a las tres columnas deseadas
nota_chilena(df_grades, "math score")
nota_chilena(df_grades, "reading score")
nota_chilena(df_grades, "writing score")
````

### 1.3.3 Reprobados y Destacados [1.5 Punto, 0.5 Cada DataFrame]

Contentos con su desempeño, la Universidad les solicita que obtengan los alumnos quienes han reprobado el año y los que han tenido desempeño de destacado. Para esto se considerará que un alumno ha reprobado el año cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerará destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una función que permita obtener tres Dataframes:
- El primero deberá poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero será un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el año.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    """Analiza los promedios de un dataframe y entrega 3 dataframes nuevos, uno
    con los alumnos reprobados, otro como el original pero indicando promedio R
    en los reprobados y otro con los alumnos destacados ordenados de mayor a
    menor.

    Parameters
    ----------
    dataframe_in : pandas.core.frame.DataFrame
        Dataframe de pandas a analizar
    """
    #Creamos un dataframe únicamente de los alumnos que tienen promedio menor a 4
    df_reprobados = dataframe_in[dataframe_in["GPA"] < 4.0]


    #Creamos un dataframe igual al inicial que cambia el valor de la columna GPA
    #por R si el promedio es menor a 4

    #Copiamos el dataframe
    df_R = dataframe_in.copy()

    #Cambiamos los promedio inferiores a 4 por R y mantenemos los demás
    df_R["GPA"] = df_R["GPA"].apply(lambda x: x if x >= 4.0 else "R")


    #Creamos un dataframe únicamente de los alumnos que tienen todas sus notas
    #mayores a 5.8
    df_destacados = dataframe_in[(dataframe_in["math score"] > 5.8) & (dataframe_in["reading score"] > 5.8) & (dataframe_in["writing score"] > 5.8)]
    #Ordenamos los promedios de mayor a menor
    df_destacados = df_destacados.sort_values(by = "GPA", ascending = False)

    #Retornamos los dataframes creados anteriormente
    return df_reprobados, df_R, df_destacados
````
